```python
#距离度量 欧氏距离
import torch

# 定义矩阵
matrix = gt_seq.squeeze(0)

# 计算矩阵中每两行之间的欧氏距离
dist_matrix = torch.cdist(matrix, matrix, p=2)

# 将欧氏距离转换为相似性度量
similarity_matrix = 1 / (1 + dist_matrix)

print("相似性矩阵:")
print(similarity_matrix)

plt.imshow(similarity_matrix.cpu().numpy())
plt.colorbar()
plt.savefig("./debug_figure/encoder/euler_distance.png")
```
